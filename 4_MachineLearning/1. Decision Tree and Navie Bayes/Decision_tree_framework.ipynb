```python
import numpy as np 
import pandas as pd 
#import graphviz
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
#from sklearn.tree import export_graphviz
from sklearn.model_selection import train_test_split
```

```python
# loading the data set
dataset = load_iris(as_frame=True)
df= pd.DataFrame(data= dataset.data)

df.head()
```

```python
# adding the target and target names to dataframe
target_zip= dict(zip(set(dataset.target), dataset.target_names))
df["target"] = dataset.target
df["target_names"] = df["target"].map(target_zip)

print(df.shape)
df.head()
```

```python
# Seperating to X and Y 
X = df.iloc[:, :4]
y = df.iloc[:, -1]
```

```python
# splitting training and test
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, shuffle=True, random_state=24)

print("X_train shape: ",X_train.shape)
print("y_train shape: ",y_train.shape)

print("X_test shape: ",X_test.shape)
print("y_test shape: ",y_test.shape)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
predictions = dt.predict(X_test)
print(predictions)
```

```python
N = 10
predicted_label = predictions[N]
predicted_label
y_test = np.array(y_test)
print("True label : ",y_test[N])
print("Predicted label : ",predicted_label)
```

```python
from sklearn.metrics import accuracy_score
print("My Implementation:\nACCURACY: {}".format(accuracy_score(y_test, predictions)))
```

```python
x1, y1 = load_img(kiwi,1)
x2, y2 = load_img(pear,2)
x3, y3 = load_img(watermelon,3)


x,y = [],[]
for cur_x,cur_y  in zip(x1,y1):
    x.append(cur_x)
    y.append(cur_y)

for cur_x,cur_y  in zip(x2,y2):
    x.append(cur_x)
    y.append(cur_y)

for cur_x,cur_y  in zip(x3,y3):
    x.append(cur_x)
    y.append(cur_y)
```

```python
X_train = np.array(X_train)
X_test =  np.array(X_test)
print("X_train : ", X_train.shape)
print("X_test : ", X_test.shape)
```

```python

model = tf.keras.applications.MobileNetV2(include_top=False, weights='imagenet', input_shape=(224, 224, 3))
predictions = model.predict(X_train)
print(predictions.shape)
```

```python
features = predictions.reshape(X_train.shape[0], -1)
print("Features shape", features.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(features, y_train)
```

```python


X_test_predictions = model.predict(X_test)
X_test_features = X_test_predictions.reshape(X_test.shape[0], -1)

predictions = dt.predict(X_test_features)
```

```python
predictions
```

```python

from sklearn.metrics import accuracy_score
print("My Implementation:\nACCURACY: {}".format(accuracy_score(y_test, predictions)))
```
