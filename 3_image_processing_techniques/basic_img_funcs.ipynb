```python
import os

root_path = os.getcwd()
img_path = root_path+"/test_imgs/test_1.png"
img_path

```

```python

import cv2
import matplotlib.pyplot as plt

bgr_image = cv2.imread(img_path)
print(type(bgr_image))
print(bgr_image.shape)
plt.imshow(bgr_image)
```

```python
rgb_image =  cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB) 
plt.imshow(rgb_image)
```

```python
red_channel = rgb_image[:,:,0]
green_channel = rgb_image[:,:,1]
blue_channel = rgb_image[:,:,2]
print(red_channel.shape)
plt.imshow(red_channel)
```

```python
print(green_channel.shape)
plt.imshow(green_channel)
```

```python
print(blue_channel.shape)
plt.imshow(blue_channel)
```

```python
gray_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2GRAY) 
  
cv2.imshow('Grayscale', gray_image) 
cv2.waitKey(0)   
  
# Window shown waits for any key pressing event 
cv2.destroyAllWindows()
```

```python
gray_blurred = cv2.blur(gray_image, (5, 5))
plt.imshow(gray_blurred)
```

### Circle detection

```python
detected_circles = cv2.HoughCircles(gray_blurred,  
                   cv2.HOUGH_GRADIENT,dp = 1, minDist  = 2, param1 = 50, 
               param2 = 50, minRadius = 1, maxRadius = 40) 
print(detected_circles)
```

```python
import numpy as np
if detected_circles is not None: 
    detected_circles = np.uint16(np.around(detected_circles)) 
    for pt in detected_circles[0, :]:
        a, b, r = pt[0], pt[1], pt[2] 
        cv2.circle(rgb_image, (a, b), r, (0, 255, 0), 2)
plt.imshow(rgb_image)
```

```python
cv2.imwrite('output.jpg', rgb_image)
```
